##Problema 1. Organizar sesiones de doblaje(I)
Se precisa coordinar el doblaje de una pel√≠cula. Los actores del doblaje deben coincidir en las
tomas en las que sus personajes aparecen juntos en las diferentes tomas. Los actores de
doblaje cobran todos la misma cantidad por cada d√≠a que deben desplazarse hasta el estudio de
grabaci√≥n independientemente del n√∫mero de tomas que se graben. No es posible grabar m√°s
de 6 tomas por d√≠a. El objetivo es planificar las sesiones por d√≠a de manera que el gasto por los
servicios de los actores de doblaje sea el menor posible. Los datos son:

N√∫mero de actores: 10

N√∫mero de tomas : 30

Actores/Tomas : https://bit.ly/36D8IuK
- 1 indica que el actor participa en la toma
- 0 en caso contrario







                                        

#Modelo

- ¬øComo implemento las restricciones?

## ¬øC√≥mo represento el espacio de soluciones?

El espacio de soluciones se representa mediante un conjunto de asignaciones de tomas a d√≠as de grabaci√≥n, cumpliendo con las restricciones del problema.

Sea $ X $ una matriz binaria de dimensi√≥n $ (30 \times D) $, donde:

- $ X_{ij} = 1 $ si la toma $ i $ est√° asignada al d√≠a $ j $.
- $ X_{ij} = 0 $ en caso contrario.
- $ D $ es el n√∫mero de d√≠as necesarios para completar la grabaci√≥n  (desconocido a priori, pero acotado por $\left\lceil \frac{30}{6} \right\rceil = 5$).

Cada soluci√≥n posible corresponde a una asignaci√≥n v√°lida de tomas a d√≠as, respetando el m√°ximo de 6 tomas por d√≠a y garantizando que los actores est√©n presentes en sus respectivas sesiones.



## ¬øCu√°l es la funci√≥n objetivo?

La funcion objetivo se define como:


$$\min \sum_{j=1}^{D} |A(j)|$$


### Donde:

- $ A(j) $  es el conjunto de actores √∫nicos que participan en las tomas asignadas al d√≠a $ùëó$.
- Es decir, se minimiza la suma total de actores involucrados en todos los d√≠as de grabaci√≥n.

## ¬øC√≥mo implemento las restricciones?

### - Cada toma debe asignarse exactamente a un d√≠a:
- Se garantiza que cada toma $ i $ pertenezca a un √∫nico d√≠a $ j $:


$$\sum_{j=1}^{D} X_{ij} = 1, \quad \forall i \in \{1, ..., 30\}$$



### - Cada d√≠a no debe tener m√°s de 6 tomas:
- Se asegura que cada d√≠a $ j $ no tenga m√°s de 6 tomas:

$$ \sum_{i=1}^{30} X_{ij} \leq 6, \quad \forall j \in \{1, ..., D\}$$


### - Si un actor tiene una toma en un d√≠a, debe ser convocado:
- $$
\text{Actor}_k \in A(j) \iff \exists\, i \in \text{Tomas del d√≠a } j \text{ tal que } A_{ik} = 1
$$


Donde $ A_{ik} $ es una matriz binaria que indica si el actor $ k $ participa en la toma $ i $.





#An√°lisis
### ¬øQue complejidad tiene el problema?. Orden de complejidad y Contabilizar el espacio de soluciones

- El problema de planificaci√≥n del doblaje descrito es combinatorio y se relaciona con variantes del bin packing problem y el problema de asignaci√≥n con restricciones, ambos NP-dif√≠ciles.

### Orden de complejidad
- El algoritmo implementado utiliza una estrategia voraz para agrupar tomas por d√≠a, orden√°ndolas por n√∫mero de actores.

- La complejidad de este enfoque puede analizarse en dos fases:


###1. Ordenamiento de tomas
  
   Ordenar 30 tomas por densidad (n√∫mero de actores involucrados):
    
$$
\mathcal{O}(n \cdot \log n) = \mathcal{O}(30 \cdot \log 30)
$$


###2. Asignaci√≥n de tomas a d√≠as

  Por cada toma se eval√∫a si cabe en el d√≠a actual (m√°ximo 6 d√≠as, m√°ximo 6 tomas por d√≠a):
$$
\mathcal{O}(n \cdot d) = \mathcal{O}(30 \cdot 5) = \mathcal{O}(150)
$$

### Complejidad total
$$
\mathcal{O}(n \cdot \log n + n \cdot d) = \mathcal{O}(n \cdot \log n)
$$

Con $ n = 30 \text{ y } d \leq 5 $


###¬øCu√°l es el tama√±o del espacio de soluciones?

El n√∫mero total de formas posibles de asignar $30$ tomas a grupos de m√°ximo 6 elementos (particiones en bloques de tama√±o $‚â§ 6$) es enorme y de orden exponencial.

Aunque calcular el valor exacto es complejo, se puede acotar el tama√±o del espacio de soluciones considerando:

N√∫mero de particiones posibles:
S = N√∫mero de formas de agrupar $30$ tomas en bloques de tama√±o $‚â§ 6$

Este n√∫mero est√° acotado por el conjunto de particiones restringidas, que se aproxima a:

Espacio de soluciones $‚àº$ n√∫mero de formas de partir $30$ elementos en grupos de $$
6 = \sum_{k=1}^{\left\lceil \frac{30}{6} \right\rceil} P(k)
$$

Donde $ùëÉ(ùëò)$ es el n√∫mero de particiones de $30$ elementos en $ùëò$ subconjuntos.




#Dise√±o
### ¬øQue t√©cnica utilizo? ¬øPor qu√©?

T√©cnica: Se ha utilizado un algoritmo voraz como t√©cnica principal de dise√±o.

### ¬øPor qu√©?

El problema de planificaci√≥n del doblaje es una variante combinatoria del problema de empaquetado (bin packing) y de la asignaci√≥n con restricciones, ambos de tipo NP-dif√≠cil, lo que hace inviable resolverlo de manera exacta en tiempo razonable con un n√∫mero creciente de tomas y actores.

El algoritmo voraz es apropiado en este contexto porque:

- Permite construir soluciones de forma iterativa, eligiendo en cada paso la mejor opci√≥n local (tomas con m√°s actores primero).

- Es r√°pido y escalable, con complejidad $ùëÇ(ùëõ‚ãÖ \log n )$.

```
# Se carga base de datos en colab
from google.colab import files
import pandas as pd
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

df = pd.read_csv(file_name, index_col=0)
df = df.iloc[1:, :10]
df.columns = [f"Actor {i}" for i in range(1, 11)]

# Se elimina filas vac√≠as
df = df.dropna(how='all')
df = df[~df.index.astype(str).str.upper().str.contains("TOTAL")]
df = df[df.sum(axis=1) > 0]

# Se convierte en enteros
df = df.astype(int)
df.index = [f"Toma {i+1}" for i in range(len(df))]

# Algoritmo voraz
tomas = df.index.tolist()
tomas_ordenadas = sorted(tomas, key=lambda t: df.loc[t].sum(), reverse=True)

dias = []

while tomas_ordenadas:
    dia_actual = []
    actores_dia = set()

    for toma in tomas_ordenadas[:]:
        if len(dia_actual) < 6:
            actores_toma = set(df.columns[df.loc[toma] == 1])
            dia_actual.append(toma)
            actores_dia.update(actores_toma)
            tomas_ordenadas.remove(toma)

    dias.append({
        "tomas": dia_actual,
        "actores": actores_dia,
        "coste": len(actores_dia)
    })

costo_total = 0

for i, dia in enumerate(dias, 1):
    print(f"D√≠a {i}:")
    print(f"  Tomas: {dia['tomas']}")
    print(f"  Actores implicados: {sorted(dia['actores'])}")
    print(f"  Coste (n¬∫ de actores): {dia['coste']}\n")
    costo_total += dia['coste']

print(f" Coste total (suma de d√≠as trabajados por actores): {costo_total}")
print(f" N√∫mero total de d√≠as utilizados: {len(dias)}")
```
