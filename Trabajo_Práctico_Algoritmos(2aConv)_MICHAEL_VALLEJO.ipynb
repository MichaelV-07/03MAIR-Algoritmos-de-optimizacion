## Â¿CÃ³mo represento el espacio de soluciones?

El espacio de soluciones se representa mediante un conjunto de asignaciones de tomas a dÃ­as de grabaciÃ³n, cumpliendo con las restricciones del problema.

Sea $ X $ una matriz binaria de dimensiÃ³n $ (30 \times D) $, donde:

- $ X_{ij} = 1 $ si la toma $ i $ estÃ¡ asignada al dÃ­a $ j $.
- $ X_{ij} = 0 $ en caso contrario.
- $ D $ es el nÃºmero de dÃ­as necesarios para completar la grabaciÃ³n  (desconocido a priori, pero acotado por $\left\lceil \frac{30}{6} \right\rceil = 5$).

Cada soluciÃ³n posible corresponde a una asignaciÃ³n vÃ¡lida de tomas a dÃ­as, respetando el mÃ¡ximo de 6 tomas por dÃ­a y garantizando que los actores estÃ©n presentes en sus respectivas sesiones.



## Â¿CuÃ¡l es la funciÃ³n objetivo?

La funcion objetivo se define como:


$$\min \sum_{j=1}^{D} |A(j)|$$


### Donde:

- $ A(j) $  es el conjunto de actores Ãºnicos que participan en las tomas asignadas al dÃ­a $ð‘—$.
- Es decir, se minimiza la suma total de actores involucrados en todos los dÃ­as de grabaciÃ³n.

## Â¿CÃ³mo implemento las restricciones?

### - Cada toma debe asignarse exactamente a un dÃ­a:
- Se garantiza que cada toma $ i $ pertenezca a un Ãºnico dÃ­a $ j $:


$$\sum_{j=1}^{D} X_{ij} = 1, \quad \forall i \in \{1, ..., 30\}$$



### - Cada dÃ­a no debe tener mÃ¡s de 6 tomas:
- Se asegura que cada dÃ­a $ j $ no tenga mÃ¡s de 6 tomas:

$$ \sum_{i=1}^{30} X_{ij} \leq 6, \quad \forall j \in \{1, ..., D\}$$


### - Si un actor tiene una toma en un dÃ­a, debe ser convocado:
- $$
\text{Actor}_k \in A(j) \iff \exists\, i \in \text{Tomas del dÃ­a } j \text{ tal que } A_{ik} = 1
$$


Donde $ A_{ik} $ es una matriz binaria que indica si el actor $ k $ participa en la toma $ i $.





#AnÃ¡lisis
### Â¿Que complejidad tiene el problema?. Orden de complejidad y Contabilizar el espacio de soluciones

- El problema de planificaciÃ³n del doblaje descrito es combinatorio y se relaciona con variantes del bin packing problem y el problema de asignaciÃ³n con restricciones, ambos NP-difÃ­ciles.

### Orden de complejidad
- El algoritmo implementado utiliza una estrategia voraz para agrupar tomas por dÃ­a, ordenÃ¡ndolas por nÃºmero de actores.

- La complejidad de este enfoque puede analizarse en dos fases:


###1. Ordenamiento de tomas
  
   Ordenar 30 tomas por densidad (nÃºmero de actores involucrados):
    
$$
\mathcal{O}(n \cdot \log n) = \mathcal{O}(30 \cdot \log 30)
$$


###2. AsignaciÃ³n de tomas a dÃ­as

  Por cada toma se evalÃºa si cabe en el dÃ­a actual (mÃ¡ximo 6 dÃ­as, mÃ¡ximo 6 tomas por dÃ­a):
$$
\mathcal{O}(n \cdot d) = \mathcal{O}(30 \cdot 5) = \mathcal{O}(150)
$$

### Complejidad total
$$
\mathcal{O}(n \cdot \log n + n \cdot d) = \mathcal{O}(n \cdot \log n)
$$

Con $ n = 30 \text{ y } d \leq 5 $


###Â¿CuÃ¡l es el tamaÃ±o del espacio de soluciones?

El nÃºmero total de formas posibles de asignar $30$ tomas a grupos de mÃ¡ximo 6 elementos (particiones en bloques de tamaÃ±o $â‰¤ 6$) es enorme y de orden exponencial.

Aunque calcular el valor exacto es complejo, se puede acotar el tamaÃ±o del espacio de soluciones considerando:

NÃºmero de particiones posibles:
S = NÃºmero de formas de agrupar $30$ tomas en bloques de tamaÃ±o $â‰¤ 6$

Este nÃºmero estÃ¡ acotado por el conjunto de particiones restringidas, que se aproxima a:

Espacio de soluciones $âˆ¼$ nÃºmero de formas de partir $30$ elementos en grupos de $$
6 = \sum_{k=1}^{\left\lceil \frac{30}{6} \right\rceil} P(k)
$$

Donde $ð‘ƒ(ð‘˜)$ es el nÃºmero de particiones de $30$ elementos en $ð‘˜$ subconjuntos.




#DiseÃ±o
### Â¿Que tÃ©cnica utilizo? Â¿Por quÃ©?

TÃ©cnica: Se ha utilizado un algoritmo voraz como tÃ©cnica principal de diseÃ±o.

### Â¿Por quÃ©?

El problema de planificaciÃ³n del doblaje es una variante combinatoria del problema de empaquetado (bin packing) y de la asignaciÃ³n con restricciones, ambos de tipo NP-difÃ­cil, lo que hace inviable resolverlo de manera exacta en tiempo razonable con un nÃºmero creciente de tomas y actores.

El algoritmo voraz es apropiado en este contexto porque:

- Permite construir soluciones de forma iterativa, eligiendo en cada paso la mejor opciÃ³n local (tomas con mÃ¡s actores primero).

- Es rÃ¡pido y escalable, con complejidad $ð‘‚(ð‘›â‹… \log n )$.

```
# Se carga base de datos en colab
from google.colab import files
import pandas as pd
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

df = pd.read_csv(file_name, index_col=0)
df = df.iloc[1:, :10]
df.columns = [f"Actor {i}" for i in range(1, 11)]

# Se elimina filas vacÃ­as
df = df.dropna(how='all')
df = df[~df.index.astype(str).str.upper().str.contains("TOTAL")]
df = df[df.sum(axis=1) > 0]

# Se convierte en enteros
df = df.astype(int)
df.index = [f"Toma {i+1}" for i in range(len(df))]

# Algoritmo voraz
tomas = df.index.tolist()
tomas_ordenadas = sorted(tomas, key=lambda t: df.loc[t].sum(), reverse=True)

dias = []

while tomas_ordenadas:
    dia_actual = []
    actores_dia = set()

    for toma in tomas_ordenadas[:]:
        if len(dia_actual) < 6:
            actores_toma = set(df.columns[df.loc[toma] == 1])
            dia_actual.append(toma)
            actores_dia.update(actores_toma)
            tomas_ordenadas.remove(toma)

    dias.append({
        "tomas": dia_actual,
        "actores": actores_dia,
        "coste": len(actores_dia)
    })

costo_total = 0

for i, dia in enumerate(dias, 1):
    print(f"DÃ­a {i}:")
    print(f"  Tomas: {dia['tomas']}")
    print(f"  Actores implicados: {sorted(dia['actores'])}")
    print(f"  Coste (nÂº de actores): {dia['coste']}\n")
    costo_total += dia['coste']

print(f" Coste total (suma de dÃ­as trabajados por actores): {costo_total}")
print(f" NÃºmero total de dÃ­as utilizados: {len(dias)}")
```
