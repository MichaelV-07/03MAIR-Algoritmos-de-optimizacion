```python
#N Reinas - Vuelta Atrás()
################################################################################

#Verifica que en la solución parcial no hay amenzas entre reinas
################################################################
def es_prometedora(SOLUCION,etapa):
################################################################
  #print(SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True

#Traduce la solución al tablero
################################################################
def escribe_solucion(S):
################################################################
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")



#Proceso principal de N-Reinas
################################################################
def reinas(N, solucion=[],etapa=0):
################################################################
### ....
  if len(solucion) == 0:         # [0,0,0...]
    solucion = [0 for i in range(N) ]

  for i in range(1, N+1):
    solucion[etapa] = i
    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

  solucion[etapa] = 0

reinas(5,solucion=[],etapa=0)
```

# Práctica individual
Problema: Encontrar los dos puntos más cercanos
Primer intento: Fuerza bruta


```python
import random
Lista_1D = [random.randrange(1, 10000) for x in range(1000)]
print (Lista_1D)
```

```python
def puntos_mas_cercanos(lista):
    min_distancia = float('inf')
    mas_cercano = None

    # Se compara cada par de elementos consecutivos en la lista
    for i in range(len(lista) - 1):
        distancia = abs(lista[i] - lista[i + 1])
        if distancia < min_distancia:
            min_distancia = distancia
            mas_cercano = (lista[i], lista[i + 1])

    return mas_cercano, min_distancia

cercanos, distminima = puntos_mas_cercanos(Lista_1D)
print(f"Puntos más cercanos: {cercanos}")
print(f"Distancia mínima: {distminima}")
```

```python
def puntos_mas_cercanos(lista):
    min_distancia = float('inf')
    mas_cercano = None

    lista.sort() # Se ordena la lista para optimizar

    # Se compara cada par de elementos consecutivos en la lista
    for i in range(len(lista) - 1):
        distancia = abs(lista[i] - lista[i + 1])
        if distancia < min_distancia:
            min_distancia = distancia
            mas_cercano = (lista[i], lista[i + 1])

    return mas_cercano, min_distancia

cercanos, distminima = puntos_mas_cercanos(Lista_1D)
print(f"Puntos más cercanos: {cercanos}")
print(f"Distancia mínima: {distminima}")
```

Segundo intento. Aplicar Divide y Vencerás


```python
import sys

# Se crea una función para encontrar la distancia mínima
def div_v(areg, izq, der):
    # Si hay solo dos elementos, devolver su distancia
    if der - izq == 1:
      return abs(areg[izq] - areg[der]), (areg[izq], areg[der])


    # Si hay solo un elemento, la distancia es infinita (no hay comparación)
    if der <= izq:
        return sys.maxsize, (None, None)

    # Se divide la lista en dos mitades
    mid = (izq + der) // 2

    # Se encuentra la distancia mínima en cada mitad
    min_izq, par_izq = div_v(areg, izq, mid)
    min_der, par_der = div_v(areg, mid + 1, der)

    # Se calcula la distancia mínima entre elementos de diferentes mitades
    min_cross = sys.maxsize
    par_cross = (None, None)
    for i in range(izq, mid + 1):
        for j in range(mid + 1, der + 1):
          dist = abs(areg[i] - areg[j])
          if dist < min_cross:
            min_cross = dist
            par_cross = (areg[i], areg[j])

     # Determinar el par con la distancia mínima
    if min_izq <= min_der and min_izq <= min_cross:
        return min_izq, par_izq
    elif min_der <= min_izq and min_der <= min_cross:
        return min_der, par_der
    else:
        return min_cross, par_cross


min_distancia, puntos_cercanos = div_v(Lista_1D, 0, len(Lista_1D) - 1)


print("La distancia mínima entre los puntos más cercanos es:", min_distancia)
print("Los puntos más cercanos son:", puntos_cercanos)
```

Calcular la complejidad. ¿Se puede mejorar?

La ecuación de recurrencia para calcular la complejidad:
$$ T(n)= 2T (n/2)+O(n^2)$$

Por lo tanto la complejidad es
 $$O(n^2)$$


 Se puede mejorar el olgoritmo ordenando la lista.


```python
# Se crea una función para ordenar
def encontrar_puntos_mas_cercanos(areg):
    areg.sort()

    min_dist = float('inf')
    puntos = (None, None)

    for i in range(len(areg) - 1):
        dist = abs(areg[i] - areg[i + 1])
        if dist < min_dist:
            min_dist = dist
            puntos = (areg[i], areg[i + 1])

    return min_dist, puntos

min_distancia, puntos_cercanos = encontrar_puntos_mas_cercanos(Lista_1D)

print("La distancia mínima entre los puntos más cercanos es:", min_distancia)
print("Los puntos más cercanos son:", puntos_cercanos)
```

# Extender el algoritmo a 2D

Fuerza Bruta


```python
Lista_2D = [(random.randrange(1, 10000),random.randrange(1, 10000)) for x in range(1000)]
print (Lista_2D)
```

```python
import math
def distancia_euclidiana(p1, p2):
    # Se calcula la distancia euclidiana entre dos puntos en un plano 2D.
    return math.sqrt((p1[0] - p2[0]) ** 2 + (p1[1] - p2[1]) ** 2)

def fuerza_bruta_puntos_cercanos(lista):
    # Se encuentra el par de puntos más cercanos usando la técnica de fuerza bruta.
    min_distancia = float('inf')
    punto_mas_cercano = None

    # Se Comparar cada par de puntos
    for i in range(len(lista)):
        for j in range(i + 1, len(lista)):
            distancia = distancia_euclidiana(lista[i], lista[j])
            if distancia < min_distancia:
                min_distancia = distancia
                punto_mas_cercano = (lista[i], lista[j])

    return punto_mas_cercano, min_distancia

# Encontrar los puntos más cercanos
puntos_cercanos, distancia_minima = fuerza_bruta_puntos_cercanos(Lista_2D)

# Mostrar los resultados
print("Puntos más cercanos:", puntos_cercanos)
print("Distancia mínima:", distancia_minima)
```

Divide y Vencerás

```python
def fuerza_bruta_puntos(lista):
    # Se encuentra el par de puntos más cercanos mediante fuerza bruta.
    min_distancia = float('inf')
    punto_mas_cercano = None
    n = len(lista)

    for i in range(n):
        for j in range(i + 1, n):
            distancia = distancia_euclidiana(lista[i], lista[j])
            if distancia < min_distancia:
                min_distancia = distancia
                punto_mas_cercano = (lista[i], lista[j])

    return punto_mas_cercano, min_distancia

def puntos_mas_cercanos_strip(strip, min_distancia):
    # Se encuentra los puntos más cercanos en una franja estrecha alrededor de la línea divisoria.
    strip.sort(key=lambda punto: punto[1])  # Ordenar por coordenada Y
    min_pair = None
    n = len(strip)

    for i in range(n):
        for j in range(i + 1, n):
            if (strip[j][1] - strip[i][1]) >= min_distancia:
                break
            distancia = distancia_euclidiana(strip[i], strip[j])
            if distancia < min_distancia:
                min_distancia = distancia
                min_pair = (strip[i], strip[j])

    return min_pair, min_distancia

def puntos_mas_cercanos_recursivo(puntos_ordenados_x):

    n = len(puntos_ordenados_x)
    if n <= 3:
        return fuerza_bruta_puntos(puntos_ordenados_x)

    mid = n // 2
    punto_medio = puntos_ordenados_x[mid]

    # Se divide en dos mitades
    izquierda = puntos_mas_cercanos_recursivo(puntos_ordenados_x[:mid])
    derecha = puntos_mas_cercanos_recursivo(puntos_ordenados_x[mid:])

    if izquierda[1] < derecha[1]:
        d_min = izquierda[1]
        mejor_par = izquierda[0]
    else:
        d_min = derecha[1]
        mejor_par = derecha[0]

    #Se crea el conjunto de puntos cercanos a la línea divisoria
    strip = [p for p in puntos_ordenados_x if abs(p[0] - punto_medio[0]) < d_min]

    # Se encuentra el par más cercano en la franja
    strip_pair, strip_distancia = puntos_mas_cercanos_strip(strip, d_min)

    if strip_distancia < d_min:
        return strip_pair, strip_distancia
    else:
        return mejor_par, d_min

def puntos_mas_cercanos(lista):
    puntos_ordenados_x = sorted(lista, key=lambda punto: punto[0])
    return puntos_mas_cercanos_recursivo(puntos_ordenados_x)

puntos_cercanos, distancia_minima = puntos_mas_cercanos(Lista_2D)

print("Puntos más cercanos:", puntos_cercanos)
print("Distancia mínima:", distancia_minima)
```

# Extender el algoritmo a 3D.
Fuerza Brura

```python
Lista_3D = [(random.randrange(1, 10000),random.randrange(1, 10000),random.randrange(1, 10000)) for x in range(1000)]
print (Lista_3D)
```

```python
def distancia_euclidiana1(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2 + (p1[2] - p2[2])**2)

def puntos_mas_cercanos_fuerzabrut(lista):
    min_distancia = float('inf')
    punto1 = None
    punto2 = None

    for i in range(len(lista)):
        for j in range(i + 1, len(lista)):
            distancia = distancia_euclidiana1(lista[i], lista[j])
            if distancia < min_distancia:
                min_distancia = distancia
                punto1, punto2 = lista[i], lista[j]

    return punto1, punto2, min_distancia

# Encontrar los puntos más cercanos con fuerza bruta
p1, p2, distancia_minima = puntos_mas_cercanos_fuerzabrut(Lista_3D)

print("Punto más cercano 1:", p1)
print("Punto más cercano 2:", p2)
print("Distancia mínima:", distancia_minima)
```

Divide y Vencerás

```python
def encontrar_puntos_mas_cercanos_divide_venceras(puntos):
    if len(puntos) == 1:
        return None, None, float('inf')

    # Se Ordena los puntos por coordenada X
    puntos.sort(key=lambda p: p[0])
    mitad = len(puntos) // 2
    punto_medio = puntos[mitad]

    # Se resolve recursivamente en cada mitad
    izq_punto1, izq_punto2, izq_distancia = encontrar_puntos_mas_cercanos_divide_venceras(puntos[:mitad])
    der_punto1, der_punto2, der_distancia = encontrar_puntos_mas_cercanos_divide_venceras(puntos[mitad:])

    # Distancia mínima entre las dos mitades
    if izq_distancia < der_distancia:
        d = izq_distancia
        punto1, punto2 = izq_punto1, izq_punto2
    else:
        d = der_distancia
        punto1, punto2 = der_punto1, der_punto2

    # Se revisa la franja cercana a la línea divisoria
    franja = [p for p in puntos if abs(p[0] - punto_medio[0]) < d]
    franja.sort(key=lambda p: p[1])  # Ordenar por coordenada Y para optimización

    for i in range(len(franja)):
        for j in range(i + 1, len(franja)):
            if (franja[j][1] - franja[i][1]) >= d:
                break
            distancia = distancia_euclidiana1(franja[i], franja[j])
            if distancia < d:
                d = distancia
                punto1, punto2 = franja[i], franja[j]

    return punto1, punto2, d

p1, p2, distancia_minima = encontrar_puntos_mas_cercanos_divide_venceras(Lista_3D)

print("Punto más cercano 1:", p1)
print("Punto más cercano 2:", p2)
print("Distancia mínima:", distancia_minima)
```
